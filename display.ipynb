```python
from IPython.display import Image, display
import json

import glob
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline


image_dir = '/research/_datasets/vist/resized/val/'
sis_file = '/research/playground/visualstorytelling-codebase/dataset/text-annotation/sis/val.story-in-sequence.json'
sis_dataset = json.load(open(sis_file, 'r'))

# Generate unique story id
list_story_id = []
for annotation in range(len(sis_dataset['annotations'])):
    story_id = sis_dataset['annotations'][annotation][0]['story_id']
    list_story_id.append(story_id)

list_story_id = list(dict.fromkeys(list_story_id))

index_story = 400

# Get story by index
listOfImageNames = []
listOfStory = []
for annotation in range(len(sis_dataset['annotations'])):
    if(sis_dataset['annotations'][annotation][0]['story_id'] == list_story_id[index_story]):
        image_file = image_dir + sis_dataset['annotations'][annotation][0]['photo_flickr_id']+'.jpg'
        listOfImageNames.append(mpimg.imread(image_file))
        listOfStory.append(sis_dataset['annotations'][annotation][0]['text'])
        
plt.figure(figsize=(20,10))
columns = 5
for i, image in enumerate(listOfImageNames):
    plt.subplot(len(listOfImageNames) / columns + 1, columns, i + 1)
    plt.imshow(image)

for i, story in enumerate(listOfStory):
    print(i+1,'.', story)
```
